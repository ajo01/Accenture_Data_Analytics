```python
import pandas as pd
```

```python
reaction_df = pd.read_csv('./data/Reactions.csv', usecols=[1,2,3])
content_df = pd.read_csv('./data/Content.csv', usecols=[1,2,3])
reaction_type_df = pd.read_csv('./data/ReactionTypes.csv', usecols=[1,2,3])
```

```python
reaction_df
```

```python
content_df
```

```python
reaction_type_df
```

```python
content_df['Category'] = content_df['Category'].str.lower()
```

```python
merge_df = pd.merge(reaction_df, content_df, on='Content ID')
merge_df
```

```python
final_df = pd.merge(merge_df, reaction_type_df, on='Type')
final_df
```

```python
final_df.to_csv('./data/output.csv', index=False)
```

```python
df = pd.read_csv('./data/output.csv')
```

```python
result = df.groupby('Category')['Score'].sum().reset_index()
result
```

```python
result = result.sort_values(by='Score', ascending=False)
result 
```
